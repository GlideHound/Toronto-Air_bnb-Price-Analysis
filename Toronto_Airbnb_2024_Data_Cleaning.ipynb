```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the csv file
df = pd.read_csv("Toronto_Airbnb_Data.csv")
df = df.replace("?", np.nan)
```

```python
# display the first five rows of the dataset to get an overview
df.head()
```

```python
# display the information of each column
df.info()
```

```python
# Remove the dollar sign from the 'price' column and convert price to float64
df['price'] = df['price'].astype(str).str.replace('[$,]', '', regex=True)
df['price'] = df['price'].astype('float')
```

```python
# replace the NaN value in price column with the calculated mean
price_avg = df['price'].mean()
df['price'] = df['price'].replace(np.nan, price_avg)
```

```python
df['price']
```

```python
# Delete rows with the NaN values in column host_is_superhost, convert it to boolean
df = df.dropna(subset=['host_is_superhost'])
df['host_is_superhost'] = df['host_is_superhost'].astype(bool)
```

```python
df['host_is_superhost']
```

```python
# Last step, save the cleaned dataframe in a csv file
df.to_csv('Cleaned_Toronto_Airbnb_Data.csv', index=False)
print("saved!")
```
